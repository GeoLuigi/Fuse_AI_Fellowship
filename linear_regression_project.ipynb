```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = open('auto-mpg.data', 'r')
fileread = file.readlines()
```

```python
data_of_data = []
for item in fileread:
    data = []
    item = item.replace("\n", "")
    item = item.replace("\t", " ")
    item = item.split(' ')
    for subitem in item:
        if subitem != "":
            data.append(subitem)
    data_of_data.append(data)
```

```python
new_data_of_data = []
for item in data_of_data:
    new_data = item[:8]
    item = ' '.join(subitem.replace('"', '') for subitem in item[8:])
    new_data.append(item)
    new_data_of_data.append(new_data)    
```

```python
print(new_data_of_data)
```

```python
file = open('auto-mpg.names', 'r')
fileread = file.read()
# fileread = fileread.split('\n')
```

```python
column_values = fileread.split('Attribute Information:')[1]
column_values = column_values.split('\n')
column_values
```

```python
columns = []
for col in column_values[2:-4]:
    strip_col = col.strip()
    split_col = strip_col.split(':')[0].split('.')[1].strip()   
    columns.append(split_col)
```

```python
print(columns)
```

```python
cars_df = pd.DataFrame(data = new_data_of_data, columns = columns)
cars_df.head()
```

```python
cars_df.to_csv('csv_data/cars_df.csv', index = False)
```

```python
cars_df.columns = ["mpg", "cyl", "disp", "hp", "wt",
                 "acc", "year", "org", "name"]
```

### Creating Regression Model for disp and mpg

```python
final_df = clean_cars_df[['mpg', 'disp', 'hp', 'wt']]
final_df.head()
```

```python
final_df.info()
```

```python
final_df = final_df.astype('float64')
final_df.info()
```

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(data = final_df, x = 'disp',
               y = 'mpg')
plt.title('Displacement Vs Miles Per Gallon')
plt.show()
```

```python
sns.kdeplot(data = final_df, x = "disp")
```

```python
data = final_df[['disp']]
labels = final_df['mpg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size = 0.2, random_state = 42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr1 = LinearRegression()
```

```python
lr1.fit(x_train, y_train)
```

```python
lr1.coef_
```

```python
lr1.intercept_
```

```python
predictions = lr1.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test, predictions)
```

```python
mean_squared_error(y_test, predictions, squared = False)
```

```python
mean_absolute_error(y_test, predictions)
```

```python
r2_score(y_test, predictions)
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(x = x_train, y = y_train, color = 'blue', label = 'training_data')
plt.scatter(x = x_test, y = y_test, color = 'green', label = 'testing_data')
plt.plot(x_test, predictions, color = 'red')
plt.title("y = {} + {}*X".format(lr1.intercept_, lr1.coef_[0]))
plt.xlabel('Displacement')
plt.ylabel("Miles Per Gallon")
plt.legend()
plt.show()
```
